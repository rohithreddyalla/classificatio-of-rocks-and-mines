# Sonar, Mines vs. Rocks

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading the dataset**

```python
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("sonar_dataset.csv",header=None)#used header = none because in dataset columns don't have any heading
df.sample(5)
```

**Exploring the dataset**

**Let's split the data into x and y**

```python
x = df.drop(60,axis='columns') 
y = df[60]
```

```python
y.sample(5)
```

```python
x.sample(5)
```

**y needs some changes to be done lets do that**

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) 
```

1 means rock , 0 means metal

```python
y.value_counts
```

**Now let's split into train and test**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(x,y,test_size=0.25, random_state= 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.sample(5)
```

```python
Y_train.sample(5)
```

```python
X_test.sample(5)
```

```python
Y_test.sample(5)
```

**Training the model**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
                          keras.layers.Dense(60,input_dim=60,activation="relu"),
                          keras.layers.Dense(30,activation="relu"),
                          keras.layers.Dense(15,activation="relu"),
                          keras.layers.Dense(1,activation="sigmoid")
])
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
model.fit(X_train,Y_train,epochs=100,batch_size=7)
```

```python
model.evaluate(X_test,Y_test)
```

Looks like there is overfitting

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
Y_test[:10]
```

**Looking at the classification report**

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(Y_test, y_pred))
```

**Now let's introduce dropout layers for handling the overfitting**

```python
modeld = keras.Sequential([
                          keras.layers.Dense(60,input_dim=60,activation="relu"),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(30,activation="relu"),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(15,activation="relu"),
                          keras.layers.Dropout(0.5),
                          keras.layers.Dense(1,activation="sigmoid")
])
modeld.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
modeld.fit(X_train,Y_train,epochs=75,batch_size=7)
```

```python
modeld.evaluate(X_test,Y_test)
```

Now the accuracy increased!!
Let's look at the classification report now

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(Y_test, y_pred))
```
